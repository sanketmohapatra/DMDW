```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
import statistics as st

df = pd.read_csv('ToyotaCorolla.csv')

df.shape

df.head()

df.sample(10)

df.info()

df.dtypes

df.describe()

df.describe(include='all')

df.isnull().sum()

df.duplicated().sum()

df = df.select_dtypes(exclude=['object'])
```

```python

df_min_max = df.copy()

# max-min normalization techniques
for column in df_min_max.columns:
	df_min_max[column] = (df_min_max[column] - df_min_max[column].min()) / (df_min_max[column].max() - df_min_max[column].min())

print(df_min_max)
```

```python
df_z_score = df.copy()

# z-score normalization techniques
for column in df_z_score.columns:
	df_z_score[column] = (df_z_score[column] - df_z_score[column].mean()) / df_z_score[column].std()

display(df_z_score)
```
